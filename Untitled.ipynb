```python
from pathlib import Path
import csv

# Source folder
source_folder = Path(r"\\HOME-NAS\media\Recipes\!ready to be filed")

# Output CSV
output_csv = Path(r"\\HOME-NAS\media\Recipes\RecipesNew.csv")

# Write the CSV (overwrites any existing file)
with output_csv.open("w", newline="", encoding="utf-8") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["Path", "Name", "Type"])

    # Recursively iterate through all files
    for file in source_folder.rglob("*"):
        if file.is_file():
            writer.writerow([
                str(file.parent),   # Directory path
                file.stem,          # Filename without extension
                file.suffix         # Extension (including the leading ".")
            ])

print(f"CSV created: {output_csv}")
```
