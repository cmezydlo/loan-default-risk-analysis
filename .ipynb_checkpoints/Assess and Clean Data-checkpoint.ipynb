### Review Application Date Data
##### NOTES
##### Peak loans in May-July 2024. 
##### Lulls in loans between Sept-Nov 2024
##### No date range is under a count of 50

```python
# Assess data distribution
g = sns.displot(kind = 'hist', data = appData,
                x = 'application_date',
                bins = 9, height = 4, aspect = 1.5)
for ax in g.axes.flat:
    ax.tick_params('x', labelrotation = 30)
```

### Review Loan Purpose Data
##### Data distributed 8-12% per category

```python
appData.loan_purpose.value_counts(normalize = True) 
```

```python
# Assess data distribution
g = sns.displot(kind = 'hist', data = appData,
                y = 'loan_purpose',
                bins = 10, height = 4, aspect = 1.5)
```

```python
# Assess counts by purpose
sns.set_style('whitegrid')
g = sns.catplot(kind = 'count', data = appData,
                y = 'loan_purpose', 
                order = appData['loan_purpose'].value_counts().index,
                height = 4, aspect = 1.5)

total = len(appData)

for ax in g.axes.flat:
    ax.set(title = 'Number of loans by purpose', ylabel = 'Purpose' )

    for bar in ax.patches:
        pct = 100 * bar.get_width() / total

        ax.text(x = bar.get_width() + 0.5,
               y = bar.get_y() + bar.get_height() / 2,
               s = f'{pct:.1f}%',
               va = 'center')

g.savefig('purpose_counts.png')
```
